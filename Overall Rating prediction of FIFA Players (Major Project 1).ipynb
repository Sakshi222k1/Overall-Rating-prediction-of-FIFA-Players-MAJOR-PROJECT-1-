```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:/Users/admin/Downloads/Telegram Desktop/data.csv")
df
```

```python
d_columns=df.columns[28:54]
df=df.drop(d_columns,axis=1)
df
```

```python
df.drop(['ID','Name','Age','Photo','Nationality','Flag','Club','Club Logo','Preferred Foot','Body Type','Real Face','Position','Jersey Number','Joined','Loaned From','Contract Valid Until','Height','Weight','Release Clause'],axis=1,inplace=True)
df=df.iloc[:,1:]
```

```python
df.drop('Work Rate',axis=1,inplace=True)
df
```

```python
#Checking if there is some null value or not
df.isnull().sum()
```

```python
#Ch3ecking data types of all the elements 
df.info()
```

```python
df1=df.iloc[:,5:]
df1.isnull().sum()
```

```python
df1.info()
```

```python
#Taking mean of columns (data type='float') having missing values 
l=[]
for i in df1.iloc[:,:]:
    a=df[i].mean()
    df[i]=df[i].replace(np.nan,a)
```

```python
df
```

```python
#Checking for null value
df.isnull().sum()
```

```python
df.info()
```

```python
# Apply label encoding in order to convert string values into numeric data type
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
df['Value_n']=label_encoder.fit_transform(df['Value'])
df['Wage_n']=label_encoder.fit_transform(df['Wage'])
```

```python
df
```

```python
df_n = df.drop(['Value','Wage'],axis='columns')
df_n
```

```python
#Checking if all the entries are of numeric data type or not
df_n.info()
```

```python
#Creating 'X' with all "independent features"
X=df_n.iloc[:,1:]
X
```

```python
#Creating 'y' with all "dependent features"
y=df_n['Overall']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=1)
```

```python
#LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
pred_linear=lin_reg.predict(X_test)
pred_linear
```

```python
#LASSO REGRESSION
from sklearn.linear_model import Lasso
lasso_reg=Lasso()
lasso_reg.fit(X_train,y_train)
pred_lasso=lasso_reg.predict(X_test)
```

```python
#RIDGE REGRESSION
from sklearn.linear_model import Ridge
ridge_reg=Ridge()
ridge_reg.fit(X_train,y_train)
pred_ridge=ridge_reg.predict(X_test)
```

```python
#ELASTIC NET REGRESSION
from sklearn.linear_model import ElasticNet
enet_reg=ElasticNet()
enet_reg.fit(X_train,y_train)
pred_elasticnet=enet_reg.predict(X_test)
```

```python
print("Linear")
linear=pd.DataFrame({'Actual':y_test,'Predicted':pred_linear})
linear
```

```python
print("Lasso")
lasso=pd.DataFrame({'Actual':y_test,'Predicted':pred_lasso})
lasso
```

```python
print("Ridge")
ridge=pd.DataFrame({'Actual':y_test,'Predicted':pred_ridge})
ridge
```

```python
print("Elastic Net")
elastic=pd.DataFrame({'Actual':y_test,'Predicted':pred_elasticnet})
elastic
```

```python
from sklearn import metrics
```

```python
#For Linear
print("LINEAR REGRESSION")
mae_linear=metrics.mean_absolute_error(y_test,pred_linear)
print("MEAN ABSOLUTE ERROR:",mae_linear)
print("ROOT MEAN SQUARED ERROR:",np.sqrt(metrics.mean_squared_error(y_test,pred_linear)))
```

```python
#For Lasso
print("LASSO REGRESSION")
mae_lasso=metrics.mean_absolute_error(y_test,pred_lasso)
print("MEAN ABSOLUTE ERROR:",mae_lasso)
print("ROOT MEAN SQUARED ERROR:",np.sqrt(metrics.mean_squared_error(y_test,pred_lasso)))
```

```python
#For Ridge
print("RIDGE REGRESSION")
mae_ridge=metrics.mean_absolute_error(y_test,pred_ridge)
print("MEAN ABSOLUTE ERROR:",mae_ridge)
print("ROOT MEAN SQUARED ERROR:",np.sqrt(metrics.mean_squared_error(y_test,pred_ridge)))
```

```python
#For Elastic Net Regression
print("ELASTIC NET REGRESSION")
mae_elastic=metrics.mean_absolute_error(y_test,pred_elasticnet)
print("MEAN ABSOLUTE ERROR:",mae_elastic)
print("ROOT MEAN SQUARED ERROR:",np.sqrt(metrics.mean_squared_error(y_test,pred_elasticnet)))
```

```python
#Plotting scatter plot of mean absolute error of different models (i.e. to compare mean absolute errors)
plt.figure(figsize=(8,5))
models=["Linear","Lasso","Ridge","ElasticNet"]
mae=[mae_linear,mae_lasso,mae_ridge,mae_elastic]
plt.plot(models,mae,'ro',linestyle='solid')
plt.grid()
plt.title("Models vs Mean absolute error")
plt.xlabel("Types of Regularization")
plt.ylabel("Mean Absolute Error")
plt.show()
```

```python
print("Therefore, for the given data least mean absolute error is shown by Linear and Ridge Regression. Hence, they are the best model for this given dataset.")
```
